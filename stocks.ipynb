```python
from matplotlib import pyplot as plt
import pandas as pd
import pandas_datareader as pdr
import seaborn as sns
```

```python
company = ["GOOGL","AMZN","TSLA","MSFT","NFLX"]

data = pdr.get_data_yahoo(company, start = "2019-01-01")["Close"]
data
```

```python
plt.figure(figsize = (12,8))
sns.set_style("darkgrid")

plt.plot(data.GOOGL)
plt.plot(data.MSFT)
plt.plot(data.AMZN)
plt.plot(data.NFLX)
plt.plot(data.TSLA)
plt.legend(loc = "best", title = "Companies", labels = ["Google","Microsoft","Amazon","Netflix","Tesla"])
```
