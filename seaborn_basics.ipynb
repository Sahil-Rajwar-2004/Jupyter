```python
import seaborn as sns
from matplotlib import pyplot as plt
```

# Loading Datasets

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planets = sns.load_dataset("planets")
```

```python
tips
```

# Scatter Plot

```python
sns.scatterplot(x = "tip", y = "total_bill", data = tips, hue = "day", size = "size", palette = "YlGnBu").set(title = "Tips")
```

# Histplot and Displot

```python
sns.histplot(tips["tip"], kde = True, bins = 15).set(title = "Tips")
```

```python
sns.displot(tips["tip"], kde = True, bins = 15).set(title = "Tips")
```

# Bar Plot

```python
sns.barplot(x = "sex", y = "tip", data = tips).set(title = "Tips")
```

# Box Plot

```python
sns.boxplot(x = "day", y = "tip", data = tips, hue = "sex").set(title = "Tips")
```

```python
sns.boxplot(x = "day", y = "total_bill", data = tips, hue = "sex").set(title = "Tips")
```

# Strip Plot

```python
sns.stripplot(x = "day", y = "tip", data = tips, hue = "sex", dodge = True)
```

# Joint Plot

```python
sns.jointplot(x = "tip", y = "total_bill", data = tips, kind = "reg")
```

```python
sns.jointplot(x = "tip", y = "total_bill", data = tips, kind = "kde", shade = True)
```

```python
sns.jointplot(x = "tip", y = "total_bill", data = tips, kind = "hex")
```

# Pair Plot

```python
titanic.corr()
```

```python
sns.pairplot(titanic.select_dtypes(["number"]), hue = "pclass")
```

# Heat Map

```python
sns.heatmap(titanic.corr(), annot = True, cmap = "icefire")
```
