```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib import style
from sklearn import linear_model
from sklearn import model_selection
from sklearn import metrics
from sklearn import preprocessing
from chemaphy import Statistics as stats
```

```python
data = pd.read_csv("E:\\Project\\DATA\\csv_files\\heart-disease-prediction.csv")
data
```

```python
e = preprocessing.LabelEncoder()
e.fit(data["Heart Disease"])
data["Heart Disease"] = e.transform(data["Heart Disease"])
```

```python
data
```

```python
data.isna().sum()
```

```python
x = data.drop(columns="Heart Disease",axis=1)
y = data["Heart Disease"]
```

```python
xTrain,xTest,yTrain,yTest = model_selection.train_test_split(x,y,test_size=0.3)
model = linear_model.LogisticRegression(max_iter=3000)
model.fit(xTrain,yTrain)
predictions = model.predict(xTest)
predictions
```

```python
m = metrics.confusion_matrix(yTest,predictions)
m
```

```python
sns.heatmap(m,annot=True)
```

```python
metrics.accuracy_score(yTest,predictions)
```

```python
stats.Accuracy(np.asarray(yTest),predictions)
```
