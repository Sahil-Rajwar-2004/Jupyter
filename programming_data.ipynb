```python
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
import pandas as pd
import seaborn as sns

sns.set_style("darkgrid")
```

```python
my_data = pd.read_csv(r"C:\Users\a\OneDrive\Desktop\Project\DATA\data\programming_lang_data.csv")
my_data
```

```python
my_data["Date"] = pd.to_datetime(my_data["Date"])
my_data.sort_values(by = "Date")
```

```python
plt.figure(figsize = (10,8))

plt.plot(my_data.Date, my_data.Python)
plt.plot(my_data.Date, my_data.C_Cpp)
plt.plot(my_data.Date, my_data.Java)

plt.xlabel("Dates")
plt.ylabel("Globally used(%)")
plt.title("Programming Language Use Globally")

plt.gcf().autofmt_xdate()
date_formate = mpl_dates.DateFormatter("%d, %b %Y")
plt.gca().xaxis.set_major_formatter(date_formate)

plt.legend(loc = "best", title = "Languages", labels = ["Python","C\C++","JAVA"])
```
